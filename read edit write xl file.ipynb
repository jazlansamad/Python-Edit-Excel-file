```python
import os
import pandas as pd

# Path to the directory containing Excel files
directory = 'C:/Users/user/file xl data anacond'

# Counter for processed files
file_count = 0

# Dictionary to optionally store DataFrames
dataframes = {}

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xls') or filename.endswith('.xlsx'):
        if filename.startswith('~$'):
            continue  # Skip temporary Excel files

        filepath = os.path.join(directory, filename)

        try:
            # Read the Excel file
            df = pd.read_excel(filepath)

            # Extract word before "university" from filename (case-insensitive)
            name_part = os.path.splitext(filename)[0]  # Remove extension
            name_lower = name_part.lower()
            if "university" in name_lower:
                before_university = name_lower.split("university")[0].strip()
            else:
                before_university = ""  # Default if "university" not found

            # Capitalize for nicer display (optional)
            before_university = before_university.title()

            # Add column 'x' and fill with extracted name
            df['Univerity'] = before_university

            # Generate short variable name from first 5 letters
            var_name = name_part[:10].lower().replace(" ", "_")

            # Store in memory
            globals()[var_name] = df
            dataframes[var_name] = df

            # Show simulated command
            print(f"{var_name} = pd.read_excel('{filepath}')")
            print(f"{var_name}.head()")
            print("-" * 50)

            # Save back to the same file
            df.to_excel(filepath, index=False)

            file_count += 1

        except Exception as e:
            print(f"Failed to read {filename}: {e}")

# Final summary
print(f"\nTotal Excel files read and updated: {file_count}")
print("Variables created:", ", ".join(dataframes.keys()))
```
